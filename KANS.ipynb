```python
import numpy as np

def solve_xor_single_layer():
    # XOR Input data
    X = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])

    # Target XOR outputs
    y_true = np.array([0, 1, 1, 0])

    # Your logic: fn(xn) = k1 * sin(k2 * xn) + k3 * xn
    # We define weights that transform the inputs into a separable space
    def fn(x, weights):
        # Using a sine transformation to capture the 'parity' of XOR
        # f(x) = sin(pi * x)
        return np.sin(np.pi * x)

    print("Target XOR Results:", y_true)
    print("-" * 30)

    results = []
    for row in X:
        x1, x2 = row

        # Applying your additive formula: bias + f1(x1) + f2(x2)
        # For XOR, if we use sin(pi*x), then:
        # sin(0) = 0, sin(pi*1) = 0. This alone won't work.
        # However, if we shift the inputs or use a different fn:

        # Let's use: f(x) = x
        # But we create a specific interaction term or a non-linear bias
        # Actually, let's use the simplest version of your fn(xn):
        f1 = np.abs(x1 - 0.5)
        f2 = np.abs(x2 - 0.5)

        # Summation + Bias
        # In this transformed space, (0,0) and (1,1) both yield 0.5 + 0.5 = 1
        # (0,1) and (1,0) both yield 0.5 + 0 = 0.5
        total = f1 + f2

        # Step function (Bias of -0.7)
        prediction = 1 if total < 0.7 else 0
        results.append(prediction)

        print(f"Input: {row} -> f1+f2: {total:.2f} -> Prediction: {prediction}")

    return np.array(results)

solve_xor_single_layer()
```

```python
import numpy as np

def fn(x):
    # Transformation: f(x) = cos(pi * x)
    return np.cos(np.pi * x)

def gaussian_activation(s):
    # Activation: e^(-s^2)
    return np.exp(-(s**2))

def solve_xor_detailed():
    inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

    print(f"{'Input (x1,x2)':<12} | {'f1':<6} | {'f2':<6} | {'Sum (s)':<8} | {'e^(-s^2)':<10} | {'Output'}")
    print("-" * 65)

    for x1, x2 in inputs:
        # Step 1: Feature Functions
        f1 = fn(x1)
        f2 = fn(x2)

        # Step 2: Summation (Aggregator)
        s = f1 + f2

        # Step 3: Activation Value
        activation_val = gaussian_activation(s)

        # Step 4: Decision Rule (Threshold)
        prediction = 1 if activation_val >= 0.5 else 0

        print(f"({x1}, {x2})        | {f1:>4.1f} | {f2:>4.1f} | {s:>8.1f} | {activation_val:>10.4f} | {prediction}")

solve_xor_detailed()
```

```python
import numpy as np

def basis_function(x, weights):
    """
    A simple Quadratic KAN edge: f(x) = w2*x^2 + w1*x + b
    This allows us to have non-zero 2nd derivatives.
    """
    w2, w1, b = weights
    return w2*(x**2) + w1*x + b

def solve_kan_newton():
    # Training Data: y = x^2
    X_train = np.linspace(-1, 1, 10)
    Y_train = X_train**2

    # Initialize KAN weights: [w2, w1, b]
    # We start with random values
    weights = np.array([0.1, 0.1, 0.1])

    print(f"{'Epoch':<8} | {'Total Loss':<12}")
    print("-" * 25)

    for epoch in range(20):
        total_loss = 0

        # We aggregate gradients and hessians over the batch
        grad_total = np.zeros(3)
        hess_total = np.zeros((3, 3))

        for i in range(len(X_train)):
            x = X_train[i]
            y_true = Y_train[i]

            # 1. Forward Pass
            y_pred = basis_function(x, weights)
            error = y_pred - y_true
            total_loss += error**2

            # 2. Jacobian (First Derivatives wrt w2, w1, b)
            # df/dw2 = x^2, df/dw1 = x, df/db = 1
            jacobian = np.array([x**2, x, 1])
            grad_total += 2 * error * jacobian

            # 3. Hessian (Second Derivatives)
            # For MSE Loss, H is approx 2 * (J^T * J)
            hess_total += 2 * np.outer(jacobian, jacobian)

        # 4. Newton Update: w = w - H^-1 * grad
        # Add small epsilon to diagonal for stability (Levenberg-Marquardt style)
        hess_total += np.eye(3) * 1e-6
        step = np.linalg.solve(hess_total, grad_total)
        weights -= step

        if epoch % 2 == 0:
            print(f"{epoch:<8} | {total_loss:<12.8f}")

    print("\nReal KAN Results (Target y = x^2):")
    print(f"Learned Function: {weights[0]:.2f}x^2 + {weights[1]:.2f}x + {weights[2]:.2f}")

solve_kan_newton()
```

```python
import numpy as np

def complex_gaussian(s, center=1 + 1j):
    # Activation: f(s) = exp(-|s - c|^2)
    dist_sq = np.abs(s - center)**2
    return np.exp(-dist_sq)

def solve_xor_newton():
    # Mapping: 0 -> 1+0j, 1 -> 0+1j
    inputs = np.array([1+0j, 0+1j])
    X = [(0,0), (0,1), (1,0), (1,1)]
    targets = np.array([0, 1, 1, 0])

    # Initialize Weights (asymmetric to avoid the symmetry trap)
    w1, w2, b = 0.6 + 0.1j, 0.1 + 0.6j, 0.2 + 0.2j

    print(f"{'Epoch':<8} | {'Total Loss':<12}")
    print("-" * 25)

    for epoch in range(10): # Notice: only 10 epochs, not 1000!
        total_loss = 0
        for i, (x1, x2) in enumerate(X):
            f1, f2 = inputs[x1], inputs[x2]

            # Forward Pass
            s = w1 * f1 + w2 * f2 + b
            pred = complex_gaussian(s)

            error = pred - targets[i]
            total_loss += error**2

            # --- The Newton Logic ---
            # 1st Derivative (Jacobian/Gradient)
            # dL/ds = 2 * error * d(pred)/ds
            # d(pred)/ds for Gaussian is -2(s-c) * pred
            g_s = -2 * (s - (1+1j)) * pred
            grad = 2 * error * g_s

            # 2nd Derivative (Hessian Approximation)
            # In Newton's method, we divide by the second derivative
            # For Gaussian, the 2nd derivative is approximately:
            hess_s = grad * (-2 * (s - (1+1j))) # Simplified Hessian

            # Update with a "Curvature-Aware" step
            # We add a small 'epsilon' (1e-5) to prevent division by zero
            step = grad / (np.abs(hess_s) + 1e-5)

            # Update weights based on their contribution to the sum
            w1 -= 0.5 * step * np.conj(f1)
            w2 -= 0.5 * step * np.conj(f2)
            b  -= 0.5 * step

        if epoch % 1 == 0:
            print(f"{epoch:<8} | {total_loss:<12.6f}")

    print("\nNewton-Step Results:")
    for x1, x2 in X:
        s = w1 * inputs[x1] + w2 * inputs[x2] + b
        print(f"Input ({x1},{x2}) -> Pred: {complex_gaussian(s):.4f}")

solve_xor_newton()
```
