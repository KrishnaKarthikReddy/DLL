```python
# ==========================================================
# ENSEMBLE CLUSTERING FOR BREAST CANCER GENE EXPRESSION
# Synthetic Gene Expression Dataset
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform

np.random.seed(42)

# ==========================================================
# 1. GENERATE SYNTHETIC GENE EXPRESSION DATA
# ==========================================================

n_samples = 200
n_genes = 1000
true_clusters = 4

X, y_true = make_blobs(
    n_samples=n_samples,
    centers=true_clusters,
    n_features=n_genes,
    cluster_std=5.0,
    random_state=42
)

# Add biological noise
noise = np.random.normal(0, 2, X.shape)
X = X + noise

# Standardize
scaler = StandardScaler()
X = scaler.fit_transform(X)

print("Dataset shape:", X.shape)

# ==========================================================
# 2. DEFINE CLUSTERING MODELS WITH HYPERPARAMETER TUNING
# ==========================================================

cluster_range = [3, 4, 5]

clusterings = []

print("\n===== Running Multiple Clustering Algorithms =====")

for k in cluster_range:

    # KMeans
    km = KMeans(n_clusters=k, n_init=20, random_state=42)
    clusterings.append(km.fit_predict(X))

    # Agglomerative
    agg = AgglomerativeClustering(n_clusters=k, linkage='ward')
    clusterings.append(agg.fit_predict(X))

    # Spectral
    spec = SpectralClustering(n_clusters=k, affinity='nearest_neighbors', random_state=42)
    clusterings.append(spec.fit_predict(X))

    # Gaussian Mixture
    gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42)
    clusterings.append(gmm.fit_predict(X))

print("Total clustering runs:", len(clusterings))

# ==========================================================
# 3. BUILD CONSENSUS MATRIX
# ==========================================================

n_runs = len(clusterings)
consensus_matrix = np.zeros((n_samples, n_samples))

for labels in clusterings:
    for i in range(n_samples):
        for j in range(n_samples):
            if labels[i] == labels[j]:
                consensus_matrix[i, j] += 1

consensus_matrix = consensus_matrix / n_runs

print("Consensus matrix built.")

# ==========================================================
# 4. FINAL CLUSTERING ON CONSENSUS MATRIX
# ==========================================================

distance_matrix = 1 - consensus_matrix
condensed_distance = squareform(distance_matrix)
linkage_matrix = linkage(condensed_distance, method='average')

final_clusters = fcluster(linkage_matrix, t=4, criterion='maxclust')

print("Final clusters formed.")

# ==========================================================
# 5. EVALUATION METRICS
# ==========================================================

sil_score = silhouette_score(X, final_clusters)
ch_score = calinski_harabasz_score(X, final_clusters)
db_score = davies_bouldin_score(X, final_clusters)

print("\n===== Evaluation Metrics =====")
print("Silhouette Score:", sil_score)
print("Calinski-Harabasz Score:", ch_score)
print("Davies-Bouldin Score:", db_score)

# ==========================================================
# 6. VISUALIZE CONSENSUS MATRIX
# ==========================================================

plt.figure(figsize=(8,6))
sns.heatmap(consensus_matrix, cmap="viridis")
plt.title("Consensus Matrix Heatmap")
plt.xlabel("Samples")
plt.ylabel("Samples")
plt.show()

# ==========================================================
# 7. COMPARE WITH SINGLE KMEANS
# ==========================================================

single_kmeans = KMeans(n_clusters=4, random_state=42)
single_labels = single_kmeans.fit_predict(X)

sil_single = silhouette_score(X, single_labels)

print("\nSilhouette Score (Single KMeans):", sil_single)
print("Silhouette Score (Ensemble Consensus):", sil_score)
```
