```python
import numpy as np

# Activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# 1. Setup Data (XOR inputs and outputs)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

# 2. Initialize Weights and Biases
np.random.seed(42) # For reproducibility
input_layer_neurons = 2
hidden_layer_neurons = 2
output_neurons = 1

# Weights between input and hidden layer
wh = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))
bh = np.random.uniform(size=(1, hidden_layer_neurons))

# Weights between hidden and output layer
wout = np.random.uniform(size=(hidden_layer_neurons, output_neurons))
bout = np.random.uniform(size=(1, output_neurons))

lr = 0.5 # Learning rate
epochs = 10000

print("--- Starting Training ---\n")

for i in range(epochs):
    # --- Forward Propogation ---
    hidden_layer_input = np.dot(X, wh) + bh
    hidden_layer_activations = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_activations, wout) + bout
    predicted_output = sigmoid(output_layer_input)

    # --- Backward Propogation ---
    error = y - predicted_output
    d_predicted_output = error * sigmoid_derivative(predicted_output)

    error_hidden_layer = d_predicted_output.dot(wout.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_activations)

    # --- Updating Weights and Biases ---
    # We show the step-by-step update for the very first iteration
    if i == 0:
        print(f"Step {i+1} Updates:")
        print(f"Initial Weights (Hidden):\n{wh}")

    # Apply Gradient Descent
    wout += hidden_layer_activations.T.dot(d_predicted_output) * lr
    bout += np.sum(d_predicted_output, axis=0, keepdims=True) * lr
    wh += X.T.dot(d_hidden_layer) * lr
    bh += np.sum(d_hidden_layer, axis=0, keepdims=True) * lr

    if i == 0:
        print(f"Updated Weights (Hidden) after step 1:\n{wh}")
        print("-" * 30)

    # Print progress every 2000 epochs
    if i % 2000 == 0:
        loss = np.mean(np.square(y - predicted_output))
        print(f"Epoch {i} Loss: {loss:.4f}")

print("\n--- Final Predictions ---")
print(predicted_output)
```

```python
import numpy as np

# Sigmoid activation and its derivative
def sigmoid(x): return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x): return x * (1 - x)

# Setup Data
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
target = np.array([[0], [1], [1], [0]])

# Initialize weights and biases
np.random.seed(42)  # For reproducibility
weights_ih = np.array([[0.5, 0.5], [0.5, 0.5]]) # Input to Hidden
weights_ho = np.array([[0.5], [0.5]])           # Hidden to Output
bias_h = np.array([0.0, 0.0])
bias_o = np.array([0.0])
lr = 0.5

# Training loop (10 Epochs)
for epoch in range(10):
    # Forward Pass
    hidden_layer_input = np.dot(inputs, weights_ih) + bias_h
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, weights_ho) + bias_o
    predicted_output = sigmoid(output_layer_input)

    # Backpropagation
    error = target - predicted_output
    d_predicted_output = error * sigmoid_derivative(predicted_output)

    error_hidden_layer = d_predicted_output.dot(weights_ho.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_output)

    # Updating Weights and Biases
    weights_ho += hidden_layer_output.T.dot(d_predicted_output) * lr
    bias_o += np.sum(d_predicted_output, axis=0) * lr
    weights_ih += inputs.T.dot(d_hidden_layer) * lr
    bias_h += np.sum(d_hidden_layer, axis=0) * lr

    # Print state at each epoch
    print(f"Epoch {epoch+1}:")
    print(f"  Weights_IH: {weights_ih.flatten()}")
    print(f"  Weights_HO: {weights_ho.flatten()}")
    print(f"  Bias_H:     {bias_h}")
    print(f"  Bias_O:     {bias_o}")
    print("-" * 30)
```

```python
import numpy as np

# 1. Activation Function & its Derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    # x is the output of the sigmoid function
    return x * (1 - x)

# 2. Setup Data
X = np.array([[0.5, 0.2]])  # Input
y_true = np.array([[1.0]])  # Target
learning_rate = 0.5

# 3. Initialize Weights & Biases (Manual values for traceability)
# Hidden Layer (2 neurons)
W_hidden = np.array([[0.1, 0.2], [0.3, 0.4]]) # 2x2 matrix
B_hidden = np.array([[0.1, 0.1]])

# Output Layer (1 neuron)
W_output = np.array([[0.2], [0.1]]) # 2x1 matrix
B_output = np.array([[0.7]])

print("--- Starting 5 Epoch Trace ---")

for epoch in range(1, 6):
    # --- FORWARD PASS ---
    # Input to Hidden
    z_h = np.dot(X, W_hidden) + B_hidden
    a_h = sigmoid(z_h)

    # Hidden to Output
    z_o = np.dot(a_h, W_output) + B_output
    y_pred = sigmoid(z_o)

    # --- CALCULATE COST (MSE) ---
    cost = 0.5 * np.power((y_true - y_pred), 2)

    # --- BACKPROPAGATION (Gradient Descent) ---
    # 1. Error at Output Layer
    error_out = y_pred - y_true
    delta_out = error_out * sigmoid_derivative(y_pred)

    # 2. Error at Hidden Layer
    error_hidden = delta_out.dot(W_output.T)
    delta_hidden = error_hidden * sigmoid_derivative(a_h)

    # --- UPDATION OF WEIGHTS & BIASES ---
    # Output Layer Updates
    W_output -= learning_rate * a_h.T.dot(delta_out)
    B_output -= learning_rate * np.sum(delta_out, axis=0, keepdims=True)

    # Hidden Layer Updates
    W_hidden -= learning_rate * X.T.dot(delta_hidden)
    B_hidden -= learning_rate * np.sum(delta_hidden, axis=0, keepdims=True)

    # --- DISPLAY PER EPOCH ---
    print(f"\nEPOCH {epoch}")
    print(f"Prediction: {y_pred[0][0]:.4f} | Cost: {cost[0][0]:.4f}")
    print(f"Hidden Weights (W1):\n{W_hidden}")
    print(f"Hidden Biases (B1): {B_hidden}")
    print(f"Output Weights (W2):\n{W_output.T}")
    print(f"Output Bias (B2): {B_output[0][0]:.4f}")
    print("-" * 30)
```
